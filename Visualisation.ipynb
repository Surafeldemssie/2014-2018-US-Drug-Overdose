```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
plt.style.use('seaborn-white')
import numpy as np
```

```python
od = pd.read_csv("ODSum.csv")
od
```

```python
od.groupby(['Year'])['totalDeath'].sum().plot.bar(color = ['red','blue'])
plt.title('2015-2018 Overdose Death')
plt.ylabel('Number of Deaths From Over Dose')
plt.xlabel('Year')
plt.gcf().set_size_inches(25, 10)
```

```python
od.groupby(['Indicator'])['totalDeath'].sum().plot.bar()
plt.title('2015-2018 Overdose Death By Drug')
plt.ylabel('Number of Deaths From Over Dose')
plt.rcParams.update({'font.size': 22})
plt.xlabel('Drug')
plt.gcf().set_size_inches(25, 10)
```

```python
sub = od.groupby(['State','Indicator'])['totalDeath'].sum()
sub_df = sub.to_frame()

```

```python
coco = od[od['Indicator'] == 'Cocaine (T40.5)']
heroin = od[od['Indicator'] =='Heroin (T40.1)']
meth =  od[od['Indicator'] == 'Methadone (T40.3)']
semi = od[od['Indicator'] == 'Natural & semi-synthetic opioids (T40.2)']
op =  od[od['Indicator'] =='Opioids (T40.0-T40.4,T40.6)']
syn = od[od['Indicator'] == 'Synthetic opioids, excl. methadone (T40.4)']
```
